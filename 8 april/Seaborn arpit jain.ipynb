## despine()

- To remove the top and right axes , since we just need x and y axis

```python
sns.set_style("white")
plt.plot(x_val,y_val)
sns.despine()
plt.show()
```

- Further customization is also possible in seaborn styles.
- We can change the parameters.
- To see the parameters available we use sns.axes_style() method.

```python
sns.axes_style()
```

```python
sns.set_style('darkgrid',{'axes.facecolor':"lightgreen"})
plt.plot(x_val,y_val)
sns.despine()
plt.show()
```

```python
sns.set_style('darkgrid',{'axes.facecolor':"lightgreen",'axes.grid':False})
plt.plot(x_val,y_val)
sns.despine()
plt.show()
```

> _changing the scale of the graph._
We use set_context() method of seaborn

 
 The availiable options are :
 
 - paper
 - notebook(Default)
 - talk

```python
sns.set_style('darkgrid',{'axes.facecolor':"lightgreen",'axes.grid':False})
sns.set_context("talk")
plt.plot(x_val,y_val)
sns.despine()
plt.show()
```

### color Palettes
- Seaborn provides a function called <code>color_palette()</code>
<code>seaborn.color_palette(palette = None, n_colors = None,desat = None)</code>
- n_colors - Number of colors in the palette, default is 6
- desat - Proportion to desaturate each color

- Readily availiable Seaborn palettes -
 - Deep
 - Muted
 - Bright
 - Pastel
 - Dark
 - Colorblind

<code>seaborn.palplot()</code> can be used to check

```python
current_palette = sns.color_palette('twilight')
sns.palplot(current_palette)
plt.show()
```

- Need to know the data carefully when to apply which palette
 - Random - Default
 - Sequential Data - Greens/Blue
 - Diverging(-1 to +1) - BrBG

```python
current_palette = sns.color_palette('BrBG',n_colors = 10)
sns.palplot(current_palette)
plt.show()
```

## Biavriate analysis

- Two variable plot
- This mainly deals with relatioship between two variables and how one variable is behaving with respect to the other

<code>jointplot()</code>

```python
df = sns.load_dataset('iris')
```

```python
sns.set_style('white')
sns.jointplot(x = 'petal_length', y = 'petal_width',data = df)
plt.show()
```

```python
sns.set_style('white')
sns.jointplot(x = 'petal_length', y = 'petal_width',data = df,hue = 'species')
plt.savefig('image.png')
plt.show()
```

- <code>kde kind in jointplot</code>

```python
# in hex plot black black region shows the most dense area(i,e where most data points are present)
sns.set_style('white')
sns.jointplot(x = 'petal_length', y = 'petal_width',data = df,kind = 'hex',color = 'black')
plt.savefig('image.png')
plt.show()
```

```python
sns.set_style('white')
sns.jointplot(x = 'petal_length', y = 'petal_width',data = df,kind = 'kde',color = 'black')
plt.savefig('image.png')
plt.show()
```

<font color = 'purple'>Relplot</font>
- <code>scatterplot() (with kind = 'scatter')(default)</code>
- <code>lineplot() (with kind = 'line')</code>

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(data=tips,x='total_bill',y='tip')
plt.show()
```

- Here we have found a fairly faded linear regression

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'line')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue = 'sex')
plt.show()
```

- This shows that males give more tips as compared to females

- Analysing differnt days separately

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue = 'sex',col='day',col_wrap = 2)
plt.show()
```

```python
sns.relplot(data=tips,x ='size',y='tip',kind = 'line',)
plt.show()
```

- From the above graph it is clear that tip increses as size increases. there is a positive correlation between them.

   We can also give confidence interval in relplot line graph by <code>ci</code> parameter.

```python
sns.relplot(data=tips,x ='size',y='tip',kind = 'line',ci = None)
plt.show()
```

___Seaborn-Plotting Categorical Data___

<code>stripplot()</code>

- stripplot() is used when one of the variable under study is categorical. It represents the data in the sorted order along any one of the axis.

```python
df=sns.load_dataset('iris')
```

```python
sns.stripplot(x='species',y='petal_length',data=df)
plt.show()
```

- This means that setosa have very smaller leaves and virginica have very larger leaves.

<code>swarmplot()</code>

```python
plt.figure(figsize = (12,6))
sns.swarmplot(x='species',y='petal_length',data=df)
plt.show()
```

___Seaborn - Distribution of Observations___

- Distribution of values within each category
<code>boxplot()</code>

__Boxplot__ is a convenient way to visualize the distribution of data through their quartiles.

one has to be categorical data and the other must be numerical data. Boxplots have lines extending from them called as whiskers. This is the reason Boxplot is sometimes reffered to as box and whiskers plot.

```python
# Anything which is beyond your whiskers is your outlier
sns.boxplot(x='species',y='petal_length',data=df,palette ='viridis')
plt.show()
```

```python
df =  sns.load_dataset('tips')
sns.boxplot(x='day',y = 'total_bill',data=df)
plt.show()
```

```python
# adding hue
df =  sns.load_dataset('tips')
sns.boxplot(x='day',y = 'total_bill',data=df,hue = 'smoker')
plt.show()
```

<code>violinplot()</code>
- __violinplot__ is a combination of boxplot with the kernel density estimation(kde).

```python
df =  sns.load_dataset('tips')
sns.violinplot(x='day',y = 'total_bill',data=df)
plt.show()
```

```python
df =  sns.load_dataset('tips')
sns.violinplot(x='day',y = 'total_bill',data=df,hue ='sex')
plt.show()
```

```python
df =  sns.load_dataset('tips')
sns.violinplot(x='day',y = 'total_bill',data=df,hue ='smoker')
plt.show()
```

```python
# if we have exactly 2 classes in hue then we can use split parameter
df =  sns.load_dataset('tips')
sns.violinplot(x='day',y = 'total_bill',data=df,hue ='sex',split = True)
plt.show()
```
